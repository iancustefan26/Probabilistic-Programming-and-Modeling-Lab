```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python

#a
lambdas = [1, 2, 5, 10]

for lambdax in lambdas:
    poisson_random = np.random.poisson(lam=lambdax, size=1000)
    plt.hist(poisson_random, bins=30, alpha=0.5, label=f'lambda={lambdax}')
```

```python
#b
result = []
for _ in range(1000):
    rand_lambda = np.random.choice(lambdas)
    poisson_random = np.random.poisson(lam=rand_lambda, size=1)
    result.append(poisson_random[0])

plt.hist(result, bins=30, color='purple', edgecolor='black', alpha=0.7)
```
