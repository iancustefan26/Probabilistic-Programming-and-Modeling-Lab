```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

**1. Variabilă aleatoare uniformă**

O variabilă aleatoare uniformă are toate valorile posibile între două limite distribuite în mod egal.

```python
# Simulăm 1000 de valori aleatoare dintr-o distribuție uniformă între 0 și 10
uniform_random = np.random.uniform(0, 10, 1000)
# binomial_var_scipy = stats.uniform.rvs(loc=0, scale=10, size=1000)

# Afișăm histograma distribuției uniforme
plt.hist(uniform_random, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.title('Distribuția Uniformă [0, 10]')
plt.xlabel('Valori')
plt.ylabel('Frecvență')
plt.show()
```

```python
print(uniform_random)
```

**2. Variabilă aleatoare binomială**

O variabilă aleatoare binomială reprezintă numărul de succese într-o secvență de experimente Bernoulli independente (de exemplu, aruncarea unei monede).

```python
# Simulăm 1000 de valori dintr-o distribuție binomială (10 aruncări de monedă, p = 0.5)
binomial_random = np.random.binomial(n=10, p=0.5, size=1000)
# binomial_var_scipy = stats.binom.rvs(n=10, p=0.5, size=1000)

# Afișăm histograma distribuției binomiale
plt.hist(binomial_random, bins=10, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribuția Binomială (n=10, p=0.5)')
plt.xlabel('Număr de succese')
plt.ylabel('Frecvență')
plt.show()
```

**3. Variabilă aleatoare exponențială**

O variabilă aleatoare exponențială este adesea folosită pentru a modela timpul între evenimente independente care au loc la o rată constantă.

```python
# Simulăm 1000 de valori aleatoare dintr-o distribuție exponențială cu rata lambda = 1/2
exponential_random = np.random.exponential(scale=5, size=1000)
# exponential_var_scipy = stats.expon.rvs(scale=2, size=1000)

# Afișăm histograma distribuției exponențiale
plt.hist(exponential_random, bins=30, color='red', edgecolor='black', alpha=0.7)
plt.title('Distribuția Exponențială (lambda = 1/2)')
plt.xlabel('Valori')
plt.ylabel('Frecvență')
plt.show()
```

**4. Variabilă aleatoare normală (Gaussiană)**

Distribuția normală este una dintre cele mai importante distribuții, folosită pentru a modela o mare varietate de fenomene.

```python
# Simulăm 1000 de valori aleatoare dintr-o distribuție normală cu media 0 și deviația standard 1
normal_random = np.random.normal(loc=0, scale=1, size=1000)
# normal_var_scipy = stats.norm.rvs(loc=0, scale=1, size=1000)

# Afișăm histograma distribuției normale
plt.hist(normal_random, bins=30, color='purple', edgecolor='black', alpha=0.7)
plt.title('Distribuția Normală (mean=0, std=1)')
plt.xlabel('Valori')
plt.ylabel('Frecvență')
plt.show()
```

```python
import arviz as az

# Generăm un interval de valori pentru axa x
x = np.linspace(-4, 4, 1000)

# Calculăm densitatea probabilității pentru fiecare valoare din x
pdf_values = stats.norm.pdf(x, loc=0, scale=1)  # Distribuție normală standard (mean=0, std=1)

# Plotăm funcția de densitate a probabilității
plt.plot(x, pdf_values, color='blue', label='Densitatea Normală (mean=0, std=1)')
az.plot_kde(normal_random, label='Densitatea obținută din eșantion')
plt.title('Funcții de Densitate')
plt.xlabel('x')
plt.ylabel('Densitate')
plt.grid(True)
plt.legend()
plt.show()
```

**Explicații:**

x = np.linspace(-4, 4, 1000): Definim un interval de valori pentru variabila aleatoare, între -4 și 4, cu 1000 de puncte.

norm.pdf(x, loc=0, scale=1): Folosim funcția pdf din scipy.stats.norm pentru a calcula funcția de densitate normală. Aceasta este o distribuție normală standard, cu media 0 și deviația standard 1.

plt.plot(): Trasează graficul funcției de densitate.
Acest grafic va afișa clasicul "clopot" al distribuției normale, centrându-se în jurul valorii medii 0, iar densitatea va scădea simetric pe măsură ce ne depărtăm de această medie.

**5. Variabilă aleatoare Poisson**

Distribuția Poisson modelează numărul de evenimente care au loc într-un interval fix de timp sau spațiu.

```python
# Simulăm 1000 de valori dintr-o distribuție Poisson cu lambda = 3
poisson_random = np.random.poisson(lam=3, size=1000)
# poisson_var_scipy = stats.poisson.rvs(mu=3, size=1000)

# Afișăm histograma distribuției Poisson
plt.hist(poisson_random, bins=15, color='orange', edgecolor='black', alpha=0.7)
plt.title('Distribuția Poisson (lambda = 3)')
plt.xlabel('Valori')
plt.ylabel('Frecvență')
plt.show()
```

```python
np.mean(poisson_random)
```

**6. Variabilă aleatoare geometrică**

Distribuția geometrică modelează numărul de încercări până la primul succes într-o serie de experimente Bernoulli.

```python
# Simulăm 1000 de valori aleatoare dintr-o distribuție geometrică cu probabilitatea de succes p = 0.3
geometric_random = np.random.geometric(p=0.3, size=1000)
# geometric_var_scipy = stats.geom.rvs(p=0.3, size=1000)

# Afișăm histograma distribuției geometrice
plt.hist(geometric_random, bins=15, color='cyan', edgecolor='black', alpha=0.7)
plt.title('Distribuția Geometrică (p=0.3)')
plt.xlabel('Număr de încercări până la primul succes')
plt.ylabel('Frecvență')
plt.show()
```
